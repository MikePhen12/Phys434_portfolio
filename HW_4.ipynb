```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import pandas as pd

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (10,10)
```

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
data_table = pd.DataFrame(data=None, index=None, columns = ['Time', 'Background_Signal', 'Data'])
data_table['Data'] = N
data_table['Time'] = t
data_table['Background_Signal'] = bkgd
print(f'The shape of the resulting table will be {data_table.shape[0]} rows by {data_table.shape[1]} columns.')
print()
data_table
```

I will be creating a scatter plot of the first 1000 points against time, then how to plot a select amount of points, and then show how to plot every tenth point within the data set.

```python
first_1000 = data_table[0:999]
# Now plotting against time
plt.scatter(first_1000['Time'], first_1000['Data'])
plt.xlabel('Time')
plt.ylabel("Data Points")
```

```python
start = 500; total_points = 2000

# I am assuming that the first data point lays at data_table[0]
range_of_points = data_table[start-1:start-1+total_points]
plt.scatter(range_of_points['Time'], range_of_points['Data'])
plt.xlabel('Time')
plt.ylabel("Data Points")
```

```python
# So to truely get every 100th point, including the first it would be 0, 99, 199, etc. So what I did was add one row then took mod 100 == 0.
# I then replaced the 'first data' point of the new data frame with the 'first data' point of the old data frame. 
data_table_hundreth = data_table.shift(periods=1)
```

```python
data_table_hundreth = data_table_hundreth[data_table_hundreth.index % 100 == 0]
data_table_hundreth.iloc[0:1] = data_table.iloc[0:1]

plt.scatter(data_table_hundreth['Time'], data_table_hundreth['Data'])
plt.xlabel('Time')
plt.ylabel("Data Points")

data_table_hundreth
```

I will now create a density plot of both time on the x-axis and signal strngth on the y-axis. There will be 10 time-bins and 25 positional bins. This is to get an accurate representation of what is going on with the data

```python
plt.hist2d(data_table['Time'], data_table['Data'], bins = (10, 25))
plt.colorbar()
plt.xlabel('Time')
plt.ylabel("Data Points")
```

The darker blue color represents times there weren't many data points collected within those areas. The opposite goes for the yellow bins; those are where there was a lot of data collected. My interpretation is that most of data was collected at the extremes of the periodic function. This would make since because since this is a sinusoidal function which tends to 'spend greater time' at the extremeties than at other places with respect to the function. 

```python
folded_data = data_table['Time'] % (2*np.pi)
data_table['Folded_Time'] = folded_data
data_table
```

```python
plt.hist2d(data_table['Folded_Time'], data_table['Data'], bins = (125, 125))
plt.colorbar()
plt.xlabel('Folded Time')
plt.ylabel("Data Points")
```

As we can see, the data is now fit into a single period because of the binning. The binning allows one to see the density of the function over a single perioud. The spots were it is more yellow leads to more data points and the blue spots lead to less data points. So at the extremes we can see there are more points which is what we expect.
