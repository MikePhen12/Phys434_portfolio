```python
plt.hist(new_data, bins=150, density=True)
plt.xlabel('Observed Signal')
plt.ylabel('Probability of Signal')
plt.title('True Signal at a value of 500')
```

When giving a true signal of 500, the plot above shows the spread of the the possible observed signal values. If I were to collect points at a true signal mean of 500, what is observed are the values and associated probabilities of the background noise being picked up when collecting the observations of the signal. 

This distribution shows that it is symmetric at approx. a observed signal of 600. But, because we are dealing with a poisson distribution, there is a slightly larger tail on the right side of the graph which makes the observed signal biased. Hence, the observed signal will alays be slightly greater than the true signal due to te fact that the background is also a positive value

```python
backgrnd = stats.poisson.rvs(loc=0, mu=100, size= 250*len(qcd_background_data['ee3']))
signal = stats.uniform.rvs(loc=0, scale=15000, size=250*len(qcd_background_data['ee3']))
new_data = backgrnd + signal

signal_edge = np.linspace(0,4000,300)
data_edge = np.linspace(0,4000,300)

## Did some research about histogram2d, this will help me plot the data
psd, x, y = np.histogram2d(new_data,signal, bins=[data_edge,signal_edge], density=True)
centered_data = (data_edge[:-1] + data_edge[1:]) / 2
signal_center = (signal_edge[:-1] + signal_edge[1:]) / 2
plt.pcolormesh(centered_data, signal_center, psd.transpose())
plt.ylabel('True signal, $P(s|d)$')
plt.xlabel('Observed data, $P(d|s)$')
```

Here, I inject true signal values starting from 0 to 15000. Now, I will be able to see differences between the observed and actual signal data. I can then create a 2d plot which will correlate the differences between the observed vs true signal data. The corresponding observed signal is about 100x greater than for every true signal value. 

```python
new_table = pd.DataFrame(columns = ['Signal', 'Background', 'Data'])
new_table['Data'] = new_data
new_table['Signal'] = signal
new_table['Background'] = backgrnd
```

```python
sig_500 = new_table.query(f'499 < Signal < 501')
plt.hist(sig_500['Data'], bins =60, density = True)
plt.xlabel('Observed Signal')
plt.ylabel('Probability of Signal')
plt.title('True Signal at a value of 500 of Observed Signal Spread')
```

This is the observed signal value spread for the true signal value at 500. This is essentially the same plot as above and just as symmetric. There is bias which offsets the histogram by about 100.

```python
obs_sig_500 = new_table.query(f'499 < Data < 501')
plt.hist(obs_sig_500['Signal'], bins =60, density = True)
plt.xlabel('True Signal')
plt.ylabel('Probability of Signal')
plt.title('Observed Signal at a value of 500 of possibility of true signal spread')
```

For an observation that yields 3500. The true signal values and the associated probabilities are surrounding the average background of 100 minus the observed signal. This would make since because in order to realize the true signal, you would want to subtract the background data from the observed signal data. 



I am trying to find the bounds that prove to be the 1$\sigma$ values for the left and right of the mean. Therefore, I need to find 34% to the left and right of mean and can find the uncertaininty of the distribution at these values. 

```python
hist, edges = np.histogram(obs_sig_500['Signal'], bins = 5000, density=True)
x=0
for i in range(0, len(edges)):
    if x < .167:
        p1 = x
        p2 = x
    else:
        print(f' {x:.5f}% and {edges[i]:.5f}')
        print(f' {x:.5f}% and {edges[-i]:.5f}')
        front_edge = edges[i]
        back_edge = edges[-i]
        break
    x = x + hist[i] + hist[-i]
print(f'1 sigma signal value at: {front_edge} and {back_edge}')
print(f'This appears at distances {front_edge-400} and {back_edge-400} from the mean')
```

Even at 1$\sigma$ values, the asymmetry of the values start to occur. The 1$\sigma$ uncertainty is $^{+30.568}_{-31.202}$ 

So, this is equal to the true signal value + 100$^{+30.568}_{-31.202}$ 

Even though that the observed signal is biased, this is good enough to approx that the signal is symmetric up to atleast 1$\sigma$ values. Hence, creating somewhat of an upper-bound limit

This makes sense though because we have a slight uncertainity in the background and it tends to be asymmetric. So we can correlate this and say that this possible true signals will also have a slight asymmetric flavor to it.
