After reviewing the links provided by the professor, I know have a good sense of the normal distribution. The normal distribution has a derived Z-table which lists how the standard deviations and the probability sums from a Gaussian distribution with a mean of 0 and a standard deviation of 1. The nice thing about Z-tables is that it gives a benchmark for the number of standard deviations related to the type of distributions. After applying PPF methods, the Z-table is used to verify the X$\sigma$ values for various Gaussian and non-Gaussian distributions. 

From doing research, it is known that Z from the Z-table = ${X - \mu \over \sigma}$, where X = value from the dataset given, $\mu$= mean value of the dataset given and $\sigma$ = standard deviation of the dataset given. 

Based on the value of Z, this also correlates to the CDF values (Area under the curve) of the Normal distributions.
 
I will integrate to nominal values of CDF values that are standard for discrete values of the Z-table"

```python
mu = 1; sigma = .01; siz=100000
distro = stats.norm.rvs(loc=mu, scale=sigma, size=siz)
fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(20,4))
    
ax1.set_title(label=f"\u03bc={mu}, \u03C3={sigma}")
ax1.hist(distro,50, density=True, color='g', alpha=.3)
ax1.tick_params(labelsize = 10)
ax1.set_xlim([mu-5*sigma,mu+5*sigma])
ax1.set_yscale('log')
x = np.linspace(mu-5*sigma,mu+5*sigma,1000)
ax1.plot(x,stats.norm.pdf(x,loc = mu, scale = sigma),linewidth = 5,alpha = 0.5)

ax2.set_title(label=f"\u03bc={mu}, \u03C3={sigma*2}")
ax2.hist(distro,50, density=True, color='g', alpha=.3)
ax2.tick_params(labelsize = 10)
ax2.set_xlim([mu-5*sigma,mu+5*sigma])
ax2.set_yscale('log')
x = np.linspace(mu-5*sigma,mu+5*sigma,1000)
ax2.plot(x,stats.norm.pdf(x,loc = mu, scale = sigma),linewidth = 5,alpha = 0.5)
   
ax3.set_title(label=f"\u03bc={mu}, \u03C3={sigma*3}")
ax3.hist(distro,50, density=True, color='g', alpha=.3)
ax3.tick_params(labelsize = 10)
ax3.set_xlim([mu-5*sigma,mu+5*sigma])
ax3.set_yscale('log')
x = np.linspace(mu-5*sigma,mu+5*sigma,1000)
ax3.plot(x,stats.norm.pdf(x,loc = mu, scale = sigma),linewidth = 5,alpha = 0.5)
              
fill_space1 = np.linspace(mu-1*sigma,mu+1*sigma,1000)
fill_space2 = np.linspace(mu-2*sigma,mu+2*sigma,1000)
fill_space3 = np.linspace(mu-3*sigma,mu+3*sigma,1000)
              
ax1.fill_between(fill_space1, stats.norm.pdf(fill_space1,loc = mu, scale = sigma), color = 'r', linewidth = 0, alpha = .9)
ax2.fill_between(fill_space2, stats.norm.pdf(fill_space2,loc = mu, scale = sigma), color = 'r', linewidth = 0, alpha = .9)
ax3.fill_between(fill_space3, stats.norm.pdf(fill_space3,loc = mu, scale = sigma), color = 'r', linewidth = 0, alpha = .9)

prob_1 = 1 - ((1 - stats.norm.cdf(mu+1*sigma, loc = mu, scale = sigma)) * 2)
prob_2 = 1 - ((1 - stats.norm.cdf(mu+2*sigma, loc = mu, scale = sigma)) * 2)
prob_3 = 1 - ((1 - stats.norm.cdf(mu+3*sigma, loc = mu, scale = sigma)) * 2)

print("      " + f'Z=1 Probability={prob_1:.5f}' + "          " + f'Z=2 Probability={prob_2:.5f}' + "            " + f'Z=3 Probability={prob_3:.5f}')
```

I explored the Rayleigh distribution.  The rayleigh distribution is a continuous probability distribution for random variables. It takes on the form of a chi-distribution with two degrees of freedom where ${f(x) = x*exp(-x^2 /2)}$.

Here I play with the scale variable. The scale is 0.2 for the top graph and 0.8 for the bottom graph. This deals with the flatness of the graph. The smaller the scale, the more steep the graph appears and the larger the scale, the more flat the graph appears. This has to do with the given sigma value of the distribution.

I will assume that I have a rayleigh like distribution for my background noise. I have decided that if I measure a value over .4 or greater using the first set of parameters will be considered signal like.

I will be studying the binomial distribution. This distribution summarizes the number of trials, or observations when each trial has the same probability when trying to obtain one particular value. This is a discrete distribution which will determine the number of successes in a sequence of n independent experiments. The equation for the binomial distribution is ${{(\frac nk)}*p^k(1-p)^{n-k} }$

The question I am trying to answer now is: What is the probability that the true noise will appear as a signal to a detector?

Defining an experimental question to follow the Binomial distribution as such: If the event being true is 60% probable within the past, If the event is to happen 50 more times, what's the probability that it happens less than 27 times?

```python
n=50; p=.6; siz=100000; 
distro = stats.binom.rvs(n,p, size=siz )
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,4))
  
ax1.set_title(f'n={n}, p={p}')
ax1.tick_params(labelsize = 10)
x = np.linspace(0,n, n+1)
ax1.set_ylabel("Probability Percentage", fontsize=12)
ax1.set_xlabel("Occurances", fontsize=12)
ax1.plot(x,stats.binom.pmf(x, n, p), 'b.', linewidth = 5,alpha = 0.5)
ax1.vlines(x, 0,stats.binom.pmf(x, n, p), 'b', linewidth = 5,alpha = 0.5)
base1 = np.linspace(0,27, 28)
ax1.fill_between(base1, stats.binom.pmf(base1,n,p), color = 'r', linewidth = 0, alpha = .9)

ax2.set_title(f'n={n}, p={p}')
ax2.tick_params(labelsize = 10)
ax2.set_yscale('log')
x = np.linspace(0, n, n+1)
ax2.set_ylabel("log(Probability Percentage)", fontsize=12)
ax2.set_xlabel("Occurances", fontsize=12)
ax2.plot(x, stats.binom.pmf(x, n, p), 'b.',linewidth = 5,alpha = 0.5)
ax2.vlines(x, 0, stats.binom.pmf(x,n,p), 'b', linewidth = 5, alpha = .5)
base2 = np.linspace(0,27, 28)
ax2.fill_between(base2, stats.binom.pmf(base2,n,p), color = 'r', linewidth = 0, alpha = .9)
prob = stats.binom.cdf(27, n, p)
z_val = (stats.norm.ppf(1-(prob)/2))
print(f'Percentage of falsely occuring signals: {prob:3f}')
print(f'Sigma value associated with percentage : {z_val:.3f}')
print()
print()
```

The discrete data and the sigma values will also produce discrete data and sigma values. This would be that there is a percentage for a false signal to occur. This probability can change quite a bit depending on the experiment and parameters chosen for the experiment

I am not surprised that there are parameters like the mean of the binomial distribution. This is because your average number of occurances may be discrete, but dividing this over the total number of occurances, n, might produce a discrete value to obtain a non-integer means. When summing over data, the distribution must be discrete, but this is not always the case when you are not summing, such as taking an average.
